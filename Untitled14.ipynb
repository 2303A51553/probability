```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

mean = 42
std_dev = 5
prob_more_than_45 = 1 - stats.norm.cdf(45, mean, std_dev)

prob_between_38_44 = stats.norm.cdf(44, mean, std_dev) - stats.norm.cdf(38, mean, std_dev)

cutoff_top_10_percent = stats.norm.ppf(0.90, mean, std_dev)

x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
y = stats.norm.pdf(x, mean, std_dev)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y, label='Normal Distribution', color='black')

x1 = np.linspace(45, mean + 4*std_dev, 1000)
y1 = stats.norm.pdf(x1, mean, std_dev)
ax.fill_between(x1, y1, color='red', alpha=0.5, label='P(X > 45)')

x2 = np.linspace(38, 44, 1000)
y2 = stats.norm.pdf(x2, mean, std_dev)
ax.fill_between(x2, y2, color='blue', alpha=0.5, label='P(38 < X < 44)')

ax.axvline(cutoff_top_10_percent, color='green', linestyle='--', label='Top 10% Cutoff')

ax.set_title('Normal Distribution of Weekly Working Hours')
ax.set_xlabel('Hours Worked')
ax.set_ylabel('Density')
ax.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"Probability of working > 45 hours/week: {prob_more_than_45:.4f}")
print(f"Probability of working between 38 and 44 hours/week: {prob_between_38_44:.4f}")
print(f"Cutoff for Top 10% workers: {cutoff_top_10_percent:.2f} hours/week")
```
